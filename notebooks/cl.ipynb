```python
df = df.fillna(df.mean())
```

```python
df = df.drop(columns=['timestamp'])
```

```python
y_tr = df['subject_id']
X = df.drop(columns=['subject_id'])
y_tr.unique()
```

```python
X.shape
```

оригинальных кластеров 8 штук

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

def cl_lin(X, n_components_after_pca, y_tr=y_tr):
    pca = PCA(n_components=n_components_after_pca, random_state=42)
    X_pca = pca.fit_transform(X)
    kmeans = KMeans(n_clusters=8, random_state=42, n_init = 10)
    labels = kmeans.fit_predict(X_pca)
    
    ari = adjusted_rand_score(y_tr, labels)
    nmi = normalized_mutual_info_score(y_tr, labels)
    print(f"ARI: {ari:.4f}")
    print(f"NMI: {nmi:.4f}")
    
    fig, axes = plt.subplots(1, 2, figsize=(25, 6))

    scatter = axes[0].scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10', s=10, alpha=0.7)
    axes[0].set_xlabel('UMAP1')
    axes[0].set_ylabel('UMAP2')
    plt.colorbar(scatter, ax=axes[0])

    scatter = axes[1].scatter(X[:, 0], X[:, 1], c=y_tr, cmap='tab10', s=10, alpha=0.7)
    axes[1].set_title(f'Main data')
    axes[1].set_xlabel('UMAP1')
    axes[1].set_ylabel('UMAP2')
    plt.colorbar(scatter, ax=axes[1])

    plt.tight_layout()
    plt.show()
    
    return labels, ari, nmi
```

```python
best_ari = 0
l = {}
best_n = 0
for n_c in [10, 20, 30, 40, 51]:
    labels, ari, nmi = cl_lin(X=X, n_components_after_pca=n_c, y_tr=y_tr)
    print(n_c)
    l[n_c] = labels
    if ari > best_ari:
        best_ari = ari
        best_n = n_c
```

```python
best_n
```

получился лучший рез без сжатия

```python
from scipy.optimize import linear_sum_assignment
from sklearn.metrics import confusion_matrix

def match_clusters(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)

    row_ind, col_ind = linear_sum_assignment(-cm)
    map1 = {}
    for i, j in zip(col_ind, row_ind):
        map1[i] = j
    y_pr = np.array([map1[x] for x in y_pred])
    
    unique_vals = np.unique(y_pr)
    map2 = {val: i+1 for i, val in enumerate(sorted(unique_vals))}
    activityID = np.array([map2[x] for x in y_pr])
    
    return activityID, map1, map2

activityID, map1, map2 = match_clusters(y_tr, l[best_n])

```

```python
subm = pd.DataFrame({
    'Index': range(0, len(activityID)),
    'activityID': activityID
})
subm.to_csv('../data/subm1.csv', index = False)
```
